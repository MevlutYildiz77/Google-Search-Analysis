```python
import pandas as pd
from pytrends.request import TrendReq, BASE_TRENDS_URL
import matplotlib.pyplot as plt
```

```python
trends = TrendReq()
trends.build_payload(kw_list=["Machine Learning"])
df = trends.interest_by_region()
df = df.sort_values(by = 'Machine Learning', ascending=False)
df = df.head(10)
print(df)
```

```python
df.reset_index().plot(x = "geoName", 
                      y = "Machine Learning",
                       figsize = (16,12), kind = "bar" )
plt.style.use("fivethirtyeight")
plt.show()
```

```python
df = TrendReq(hl= 'en-US', tz= 720)
df.build_payload(kw_list=['Machine Learning'])
df = df.interest_over_time()
fig, ax = plt.subplots(figsize = (15, 12))
df["Machine Learning"].plot()
plt.style.use("fivethirtyeight")
plt.title("Total Search for Machine Learning")
plt.xlabel('Year')
plt.ylabel("Total Count")
plt.show()
```
